```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
work_with_data['Revenue'] = work_with_data['Revenue'].str[2:]
work_with_data['Revenue'] = work_with_data['Revenue'].str.replace(',','')
work_with_data['Revenue'] = pd.to_numeric(work_with_data['Revenue'])
```

```python
work_with_data
```

```python
work_with_data['Total assets'] = work_with_data['Total assets'].str[2:]
work_with_data['Total assets'] = work_with_data['Total assets'].str.replace(',','')
work_with_data['Total assets'] = pd.to_numeric(work_with_data['Total assets'])
```

```python
work_with_data[['Revenue', 'Total assets']].plot.bar(figsize=(15,5), title= 'Gains')
```

```python
work_with_data[['Revenue', 'Total assets']].plot(figsize=(15,5))
```

```python
# Check null or missing values
work_with_data.isna().any()
```

```python
# In this dataset(data frame) is not useful for us therefor we need to remove this column from dataset
del work_with_data['Source']
work_with_data
```

```python
work_with_data['Year'] = work_with_data['Year'].str[-4:] # verilmis araliqdaki illerden sonuncusunu goturmek isteyirem ve bu columndaki datalarin string oldugunu nezere alarsaq indexing emeliyyatini tetbiq etmek olar 
# indexingde sondan saymaga baslayir ve -4 olur lakin datanin tipi helede integer deyil 
work_with_data
```
